Setup matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

Read the image

```python
img = mpimg.imread('data/stinkbug.png')
# print(img)
```

Plotting numpy array as images

```python
imgplot = plt.imshow(img)
```

```python
# Get the luminosity of the image, not that these transforms the 3-dimensional array W x H x RGB to W x H.
# This works because the image is a gray scale, so we don't need the last dimension (RGB).
lum_img = img[:,:,0]
# print(img.shape)
# print(lum_img.shape)
```

Plot the luminosity image. Using the default 'viridis' color map. See here for a complete list of color maps: https://matplotlib.org/examples/color/colormaps_reference.html

```python
plt.imshow(lum_img)
```

```python
plt.imshow(lum_img, cmap='hot')
```

```python
imgplot = plt.imshow(lum_img)
imgplot.set_cmap('nipy_spectral')
plt.colorbar()
```

## Numpy ravel()

This function flattens multi dimensional array to 1-D array.

```python
# Flattening 2-D array
a = np.array([[1, 2], [3, 4]])
print(a.ravel())

# Flattening 3-D array
b = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print(b.ravel())
```

## Histogram of Image

```python
print(lum_img.ravel())
plt.hist(lum_img.ravel(), bins=256, range=(0.0, 1.0), fc='k', ec='k')
```

```python
f, axarr = plt.subplots(2)
axarr[0].imshow(lum_img)
axarr[1].imshow(lum_img, clim=(0.0, 0.7))
```

## Array Interpolation Schemes

```python
from PIL import Image
img = Image.open('data/stinkbug.png')
img.thumbnail((64, 64), Image.ANTIALIAS)
imgplot = plt.imshow(img)
```

```python
imgplot = plt.imshow(img, interpolation='nearest')
```

```python
imgplot = plt.imshow(img, interpolation='bicubic')
```
